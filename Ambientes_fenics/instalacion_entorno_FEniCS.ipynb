## Instalación a través de conda (Linux y MacOS)

__NOTA__: La instalacion de nuestro entorno de desarrollo con fenics requiere de dependencias que solo estan disponibles en UNIX por lo que la instalacion mediante el entono CONDA o ANACONDA solo sera posible en Linux y MacOS

![image](https://www.anaconda.com/wp-content/uploads/2022/12/anaconda_secondary_logo.svg)

En un nuevo terminal CTRL+ALT+T,procedemos a descargar e instlaar instalar Miniconda3:



````python
wget https://repo.continuum.io/miniconda/Miniconda3-latest-Linux-x86_64.sh -O miniconda.sh
bash miniconda.s
````

Definir el directorio de destino de la instalación. Luego de ello, ejecutar:

```
$ source .bashrc

```

Dentro del entorno conda, instalar Fenics:

```

$ conda create -n fenics -c conda-forge fenics

```

Screenshot from 2023-09-25 12-08-12.png

Probar la instalación con:
```
$ python -c "import fenics"
```

```
$ conda install -c conda-forge scipy ipython meshio
```

```
$ touch my_env.yml
```

### Instalación de Docker en Windows con WSL 2




Docker en Windows puede ser potenciado con Windows Subsystem for Linux 2 (WSL 2) para obtener un rendimiento mejorado y una integración más estrecha. A continuación, se presentan los pasos para instalar Docker en Windows con WSL 2.

##### Requisitos previos:

- Asegúrate de tener Windows 10 Pro, Enterprise o Education (versión 1903 o posterior) con la funcionalidad Hyper-V habilitada.

- Verifica que la virtualización esté habilitada en la configuración de la BIOS/UEFI de tu sistema.

#### Pasos para la instalación:

##### 1. Habilita WSL:

- Escriba PowerShell en el buscador de Windows, ábralo como Administrador y ejecute el siguiente comando para habilitar la característica de WSL:

```bash
dism.exe /online /enable-feature /featurename:Microsoft-Windows-Subsystem-Linux /all /norestart
```

##### 2. Habilita la característica Hyper-V:

- Ejecuta el siguiente comando en PowerShell como administrador para habilitar la característica Hyper-V:

```bash
dism.exe /online /enable-feature /featurename:VirtualMachinePlatform /all /norestart
```

- Reinicia tu computadora si se te solicita.


##### 3. Configura WSL 2 como versión predeterminada:

- Abre PowerShell como administrador y ejecuta el siguiente comando para configurar WSL 2 como la versión predeterminada:

```bash
wsl --set-default-version 2
```
##### 4. Descarga e instala Docker Desktop:

Visita el [sitio oficial de Docker Desktop](https://www.docker.com/products/docker-desktop) y descarga el instalador para Windows.
##### 5. Instala Docker Desktop:

Ejecuta el instalador descargado y sigue los pasos del asistente de instalación. Asegúrate de seleccionar la opción para utilizar WSL 2 durante la instalación como se muestra en el paso 3.

Al finalizar este proceso deberá poder tener acceso a una interfaz gráfica como la siguiente

![image.png](https://blog.appseed.us/content/images/2022/01/image.png)


##### 6. Descarga una distribución de Linux desde la Microsoft Store:

- Abre la Microsoft Store, busca una distribución de Linux (por ejemplo, Ubuntu) y descárgala. Sigue las instrucciones para configurar tu distribución.

##### 7. Verifica la instalación:

Abre la línea de comandos de tu distribución de Linux en WSL e ejecuta el siguiente comando para verificar que Docker esté funcionando:

```bash
docker --version
```

Este comando debería mostrar la versión de Docker instalada.

##### 8. Ejecuta un contenedor de prueba:

Abre la línea de comandos y ejecuta el siguiente comando para descargar e iniciar un contenedor de prueba:

```bash
docker run hello-world
```

Si todo está configurado correctamente, verás un mensaje indicando que Docker está funcionando y ha ejecutado un contenedor de prueba.

### Instalación de Docker en macOS



Docker permite crear y gestionar contenedores de manera eficiente en sistemas operativos macOS. A continuación, se detallan los pasos para instalar Docker en tu Mac.

#### Pasos para la instalación:

Docker Desktop utiliza HyperKit como backend para ejecutar contenedores. Si necesitas personalizar la configuración de HyperKit, puedes hacerlo mediante la edición del archivo de configuración. Por ejemplo, puedes encontrar el archivo en la ruta:

### 1. Descarga la imagen de Docker

### 6. Creación del contenedor
A partir de una misma imagen de Docker podemos montar diversos contenedores que están aislados entre sí y que pueden ser de utilidad cuando se está desarrollando aplicaciones de ingeniería donde se requiere compatibilidad entre librerías para esto muy someramente explicaremos las diferentes opciones que tenemos al momento de crear nuestro contenedor



- para poder montar nuestro contenedor será esencial hacer uso del comando



```

docker run -d  sebcastillo/ciaid_fenics_env:release  -{opciones}

```

### 7. Opciones del contenedor


#### Usar ficheros de nuestro sistema operativo dentro del contenedor.

Comúnmente requeriremos de llamar archivos que se alojen dentro de una carpeta del sistema operativo para ejecutar algún caso de simulación con lo que podemos estableces acceso al contenedor a una ruta específica de nuestra máquina. Por ejemplo para usar todos los ficheros a partir de la ruta en la que ejecutemos nuestro montaje de el contenedor podremos usar el siguiente comando.


__En Windows__
```

 docker run  -ti -v ${pwd}:/home/fenics/{folder dentro del contenedor} docker.io/sebcastillo/ciaid_fenics_env:release

```
__En Linux o MacOS__

```

 docker run  -ti -v $(pwd):/home/fenics/{folder dentro del contenedor} docker.io/sebcastillo/ciaid_fenics_env:release

```


- Habilitar puertos y local host. Esta utilidad suele ser usada por los desarrolladores web para el desarrollo de aplicativos en los contenedores para nosotros será útil para poder ejecutar cuadernillos de Jupyter.



```

 docker run -d -p 127.0.0.1:8888:8888 docker.io/sebcastillo/ciaid_fenics_env:release

```

- Nombrar nuestro contenedor. Esta opción es útil para poder tener control de nuestros contenedores





```

docker run .ti --name {nombre del contenedor} docker.io/sebcastillo/ciaid_fenics_env:release

```





__Comando para montaje de contenedor de FEniCS con uso de Jupyter Notebook___



```

docker run --name fenics-notebook -w /home/fenics -v ${pwd}:/home/fenics/shared -d -p 127.0.0.1:8888:8888 docker.io/sebcastillo/ciaid_fenics_env:release 'jupyter-notebook --ip=0.0.0.0'
```
